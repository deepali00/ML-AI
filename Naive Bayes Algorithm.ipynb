```python
import pandas as pd
```

```python
dataset= pd.read_csv('sms.tsv',sep='\t',names=['Label','Messages'])
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset['Label']= dataset['Label'].map({'ham':0,'spam':1})
```

```python
dataset.head()
```

```python
import re
```

```python
def clean(x):
    s= re.sub('[^A-Za-z]',' ',x)
    s= re.sub('\s+',' ',s)
    s= s.strip()
    return s.lower()
```

```python
dataset['Messages']= dataset.Messages.apply(clean)
```

```python
dataset.head()
```

```python
x= dataset.Messages.values
y= dataset.Label.values
```

```python
from sklearn.model_selection import train_test_split 
```

```python
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=.3, random_state=13)
```

```python
from nltk.corpus import stopwords
```

```python
stopwords= set(stopwords.words('english'))
```

```python
stopwords
```

```python
if 'not' in stopwords:
    stopwords.remove('not')
```

```python
if 'not'in stopwords:
    print('found')
else:
    print('not found')
```

```python
### data--->numeric form
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv1= CountVectorizer(stop_words=stopwords)
```

```python
cv_train= cv1.fit_transform(xtrain).toarray()

cv_test= cv1.transform(xtest).toarray()
```

```python
cv_train.shape
```

```python
cv_test.shape
```

```python
cv1.get_feature_names()
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
nb= MultinomialNB()
nb.fit(cv_train,ytrain)
```

```python
test_score= nb.score(cv_test,ytest)
```
