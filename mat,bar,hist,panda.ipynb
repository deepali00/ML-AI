```python
###   HOW TO FIND MISSING DATA
df1=pd.read_csv('train.csv')
df1.head()
```

```python
df1.isnull()    #gives the null value of whole table
```

```python
df1[df1.Cabin.isnull()] #gives null value of cabin
```

```python
df1.isnull().sum()  # displayes missing data column wise
```

```python
## fill those null values
df1.Age.aggregate(['min','max','mean','median'])
```

```python
df1.fillna({'Age':df1.Age.median()},inplace=True)
```

```python
df1.isnull().sum()
```

```python
## drop the missing data
df1.dropna(inplace=True)
```

```python
df1.isnull().sum()
```

```python
df1.head()
```

```python
#reset indexing
df1.reset_index()
```

```python
df1.reset_index(drop=True,inplace=True)
```
